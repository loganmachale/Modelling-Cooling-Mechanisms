```python
"""Code for plotting temperature over time for each of the 3 rods during the cooling process"""

import matplotlib.pyplot as plt
import numpy as np


fname = 'experimentdata.csv'
x_name = 'Frequency'
x_units = 'Hz'
y_name = 'Vout/Vin'
y_units = 'no units'

titles = ("Temperature over time for rough rod while cooling",
          "Temperature over time for polished rod while cooling",
          "Temperature over time for lacquered rod while cooling")

colours = ["r.", "g.", "b."]

data = np.loadtxt(fname, delimiter=',', comments='#',
                  usecols=(0, 4, 5, 6))
# access the data columns and assign variables x,y  and  uncetainty in y y_sigma
x = data[:, 0]
for i in range(1, 4):
    y = data[:, i]
    # y_sigma = data[:, ]
    plt.figure(1)
    plt.errorbar(x, y, yerr=2.5, xerr=1, fmt=colours[i-1])
    plt.title(titles[i-1])
    plt.xlabel("Time(s)")
    plt.ylabel("Temperature(K)")
    plt.show()

plt.figure(2)
plt.plot(x, data[:, 1], colours[0], x, data[:, 2],
         colours[1], x, data[:, 3], colours[2])
plt.title("Temperature over time for all 3 rods")
plt.legend(["Rough", "Polished", "Lacquered"])
plt.xlabel("Time(s)")
plt.ylabel("Temperature(K)")
plt.show()
```
